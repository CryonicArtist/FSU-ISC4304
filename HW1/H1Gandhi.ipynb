```python
# Question 4
import random

random_numbers = []
total_sum = 0

for i in range(10):
    num = random.random()
    random_numbers.append(num)
    total_sum += num

print("List of random numbers:", random_numbers)
print("Total Sum:", total_sum)
```

```python
# Question 5
def calculate_series(K):
    total = 0
    for k in range(1, K + 1):
        total += 1 / k
    return total

print("K=10:", calculate_series(10))
print("K=1000:", calculate_series(1000))
print("K=10000:", calculate_series(10000))
```
