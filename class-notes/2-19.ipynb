```python
import numpy as np
r = np.random.rand(1000,1000)
```

```python
def test(r):
    s = []
    for i in range(1000):
        for j in range(1000):
            s += r[i,j]
    return s

def numpytest(r):
    return np.sum(r)
```

```python
import time

start = time.perf_counter()
test(r)
end = time.perf_counter()
print(end-start)
```

```python
start = time.perf_counter()
s = numpytest(r)
end = time.perf_counter()
print(end-start)
```

```python
def expression(x,y):
    z = np.cos(x) * np.sin(y*y)
    return z
```

```python
l1 = np.random.rand(100000)
tic = time.perf_counter()
z = [expression(li,li+2) for li in l1]
toc = time.perf_counter()
loop = toc-tic
print("loop", loop)

tic = time.perf_counter()
ln1 = np.array(l1, dtype=float)
z = expression(ln1,ln1+2)
toc = time.perf_counter()
nump = toc - tic
print('vector:', nump)
print('speedup:', loop/nump)
```
